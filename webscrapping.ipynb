```python
###### MAIN

import pandas as pd 
import numpy as np
det_indx= {'unit_num': 13,
           'street': 15,
           'city': 19,
           'prov': 23,
           'district': 35,
           'latitude': 42,
           'longitude': 44,
           'sold_date': 51,
           'dom':55,
           'sold_price': 63,
           'listing_price': 67,
           'beds': 75,
           'baths': 79,
           'size': 83,
           'mls': 91,
           'brokerage': 99}

columns = ['unit_num', 'street', 'city', 'prov', 'district', 'latitude', 'longitude', 
           'sold_date', 'dom', 'sold_price','listing_price', 'beds', 'baths', 'size', 'mls', 'brokerage']

prop_data = pd.DataFrame(columns=columns)

for prop in zolo_prop:
    if "MLS" in prop:
        row = {}
        details = prop.split("&")

        unit_num = details[13].split(';')[1]

        street = details[15].split(';')[1]

        city = details[19].split(';')[1]

        prov = details[23].split(';')[1]

        district = details[35].split(';')[1]

        #latitude
        latitude = details[42].split('=')[1].split()[0][1:-1]


        #longitude
        longitude = details[44].split('=')[1].split()[0][1:-1]

        # sold date
        sold_date = details[51].split('•')[1]

        # days on market
        dom = details[55].split('"p1">')[1]#.split()[0]

        # sold price
        sold_price = details[63].split('$')[1]#.strip()

        # listing price
        listing_price = details[67].split('$')[1]#.strip()

        beds = details[75].split(';')[1]
#         print(beds)
#         print()

        baths = details[79].split(';')[1]
#         print(baths)
#         print()

        # size info may not be available, subsequent detail index will change
        # check if size detail exists in listing, 600-699 sqft or 1500-2000 sqft
        if ('99 sqft' in prop) or ('00 sqft' in prop):
            size = details[83].split(';')[1]
            mls = details[91].split(';')[1]
            brokerage = details[99].split(';')[1]
            
        else:

            size = np.NaN
            mls = details[87].split(';')[1]
            brokerage = details[95].split(';')[1]


        row = [unit_num, street, city, prov, district, latitude, longitude, 
               sold_date, dom, sold_price, listing_price, beds, baths, size, mls, brokerage]
        
        row = pd.DataFrame([row], columns = columns)


#         print(row)
            
        prop_data = pd.concat([prop_data, row])
        

# remove all unwanted spaces in each columns
def strip_string(value):
    # Check if the value is NaN (None) and return it as is
    if pd.isna(value):
        return value
    # Otherwise, apply the strip() method to remove leading and trailing whitespaces
    return value.strip()

prop_data = prop_data.applymap(strip_string)   

prop_data = prop_data.reset_index(drop=True)

prop_data
```

```python
print(prop_data[['beds','baths']].head())

# remove the bd and ba suffixes
prop_data['beds'] = prop_data['beds'].apply(lambda x: x.split()[0])
prop_data['baths'] = prop_data['baths'].apply(lambda x: x.split()[0])

prop_data[['beds','baths']].head()
```

```python
print(prop_data['sold_date'].head())

# clean sold_date
def get_sold_date(date, currentYear=2023):
    months_list = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct']
    month = date.split()[0]
    month_num = months_list.index(month)
    
    date_num = date.split()[1].split('th')[0]
    return pd.to_datetime(f'{currentYear}-{month_num+1}-{date_num}')

prop_data['sold_date'] = prop_data['sold_date'].apply(lambda x: get_sold_date(x))

prop_data['sold_date'].head()
```

```python
print(prop_data['dom'].head())

# clean days on market
prop_data['dom'] = prop_data['dom'].apply(lambda x: x.split()[0])
prop_data['dom'] = pd.to_numeric(prop_data['dom'])

prop_data['dom'].head()
```

```python
print(prop_data[['sold_date','dom']].head())

# get listing date
listing_date = prop_data['sold_date'] - pd.to_timedelta(prop_data['dom'], unit='days')

# add to main dataset
prop_data['listing_date'] = listing_date

listing_date.head()
```

```python
print(prop_data['size'].head())

# clean unit sizes
size_range = prop_data['size'].apply( lambda x: x.split()[0] if type(x) == str else x)

# get average size of range
size = size_range.apply(lambda x: ( int(x.split('-')[0]) + int(x.split('-')[1]) ) // 2 if type(x) == str else x )

# round off size to end in 00.0, ie. 500
size = size.apply(lambda x: x + 1 if x%2 == 1 else x)
size = size.astype('Int64')

prop_data['size'] = size

prop_data['size'].head()
```

```python
def get_coordinates(address):
    # Initialize a geocoder with the Nominatim service
    geolocator = Nominatim(user_agent="forward_geocoder")

    # Use the geocoder to get the location details (longitude and latitude)
    location = geolocator.geocode(address)

    if location:
        latitude = location.latitude
        longitude = location.longitude
        return latitude, longitude
    else:
        return None, None
```

```python
# find new indexes with the new split

def get_prop_detail_index(
    n=None, 
    row=None,
    cols=columns):
    """
    This function finds the new index integers for each detail from the list called columns.
    n is an integer for the row we want to check the index
    
    row is a list of property details in the order of columns
    
    cols is a list of columns defined above
    """
    print(columns)

    first_prop_index = {}
    first_prop = zolo_prop[n].split('&')

    for col_num, det in enumerate(row):
        first_prop_index[cols[col_num]] = []
        for i,vv in enumerate(first_prop):
            if det in vv:
                first_prop_index[cols[col_num]].append(i)
    #             print(f"{cols[col_num]} has index {i}")
    return first_prop_index            
```

```python
import pandas as pd
from datetime import datetime

col_names = ["sold_price", "mls_num", "address", "bedrooms", "bathrooms","parking", "brokerage", "scrape_date"]

property_data = pd.DataFrame(columns = col_names)


for prop in properties:
    
    if 'CAD' in prop:
        # sold prices
        price = prop.split(detail_names[0])[1].split()[0]

        # MLS num
        mls = prop.split(detail_names[1])[1].split()[0]

        # address
        address = prop.split(detail_names[2])[1].split('&lt;')[0]

        # bedrooms
        beds = prop.split(detail_names[3])[1].split('&lt;')[0]

        # bathrooms
        baths = prop.split(detail_names[3])[2].split('&lt;')[0]

        # parking
        parking = prop.split(detail_names[3])[3].split('&lt;')[0]

        # brokerage
        brokerage = prop.split(detail_names[4])[1].split('&lt;')[0]
        
        # scraping date
        date = datetime.today().strftime('%Y-%m-%d')

        row = pd.DataFrame([[price, mls, address, beds, baths, parking, brokerage, date]], columns =col_names)

        property_data = pd.concat([property_data, row])
```
